```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
```

```python
path_fish = r"D:\Python_first\learn\pytorch\data\fish_\train"
path_mask = r"D:\Python_first\learn\pytorch\data\mask_image"
```

```python
def makelst(file1):
    path1_fish = path_fish + '\\' + file1
    path1_mask = path_mask + '\mask_' + file1[-2: ]
    l = []
    for file2 in os.listdir(path1_fish):
        path2_fish = path1_fish + '\\' + file2
        path2_mask = path1_mask + '\mask_' + file2[5:]
        l.append((path2_fish, path2_mask))
    return l
```

```python
def merge(imgPenguin, imgPenguinMask, imgHiking):
    num = random.randint(6, 10) / 10
    gao = int(imgHiking.shape[0]*num)
    chang = int(imgHiking.shape[1]*num)
    
    imgHiking = cv2.resize(imgHiking, (chang, gao))
    ratio = 1
    height = imgPenguinMask.shape[0]
    length = imgPenguinMask.shape[1]
    imgPenguinMask = cv2.resize(imgPenguinMask, (int(length*ratio), int(height*ratio)))
    top = (height-int(height*ratio)) // 2
    bottom = height-int(height*ratio) - top
    left = (length - int(length*ratio)) // 2
    right = length - int(length*ratio) - left

    imgPenguinMask = cv2.copyMakeBorder(imgPenguinMask, top, bottom, left, right, cv2.BORDER_CONSTANT,value=[0,0,0])
    # 高斯模糊
    imgPenguinMask=cv2.GaussianBlur(imgPenguinMask, (5, 5), -1)
#     plt.imshow(imgPenguinMask)
#     plt.show()
    
    imgPenguinMaskNorm = imgPenguinMask.astype(np.float) / imgPenguinMask.max()
    outputImage = imgHiking.copy()
    
    targetTop = random.randint(0, (gao-height) if (gao-height) > 0 else 0)
    targetLeft = random.randint(0, (chang-length) if (chang-length) > 0 else 0)
    
    outputImage[targetTop:targetTop+imgPenguin.shape[0], targetLeft:targetLeft+imgPenguin.shape[1]] = imgPenguinMaskNorm * imgPenguin + (1 - imgPenguinMaskNorm) * outputImage[targetTop:targetTop+imgPenguin.shape[0], targetLeft:targetLeft+imgPenguin.shape[1]]

    # 显示图片
#     plt.figure(figsize=(6,6))
#     plt.imshow(outputImage)
#     plt.show()

    return outputImage[targetTop:targetTop+imgPenguin.shape[0], targetLeft:targetLeft+imgPenguin.shape[1]]
```

```python
bg_path = r'./bg5.png'
save_ = r'\\fish5_'
for file1 in os.listdir(path_fish):
    lst = os.listdir(path_fish + '\\' + file1)
    
    if len(lst) < 100:
        pic_list = makelst(file1)
        for pic, mask in pic_list:
            print(pic)
            imgPenguin=np.array(cv2.imread(pic))[:, :, ::-1]
            imgPenguinMask=np.array(cv2.imread(mask))[:, :, ::-1]
            imgHiking = np.array(cv2.imread(bg_path))[:, :, ::-1]
            
            try:
                merge_pic = merge(imgPenguin, imgPenguinMask, imgHiking)
                
                print('---')
                path_save = r'D:\Python_first\learn\pytorch\data\fish_increase\train\\' + file1 + save_ + pic[-22:]
                cv2.imwrite(path_save, merge_pic[:, :, ::-1])
                
                # 显示图片
#                 plt.figure(figsize=(6,6))
#                 plt.imshow(merge_pic)
#                 plt.savefig(path_save)
#                 plt.show()
                
            except Exception as e:
                print(e)
                continue
            
            
    elif len(lst) < 1000:
#         print(file1)
        pic_list = makelst(file1)
#         print(pic_list)
        
        for pic, mask in pic_list:
            print(pic)
            imgPenguin=np.array(cv2.imread(pic))[:, :, ::-1]
            imgPenguinMask=np.array(cv2.imread(mask))[:, :, ::-1]
            imgHiking = np.array(cv2.imread(bg_path))[:, :, ::-1]
            
            try:
                merge_pic = merge(imgPenguin, imgPenguinMask, imgHiking)
                
                print('---')
                path_save = r'D:\Python_first\learn\pytorch\data\fish_increase\train\\' + file1 + save_ + pic[-22:]
                cv2.imwrite(path_save, merge_pic[:, :, ::-1])
                
                # 显示图片
#                 plt.figure(figsize=(6,6))
#                 plt.imshow(merge_pic)
#                 plt.savefig(path_save)
#                 plt.show()
                
            except Exception as e:
                print(e)
                continue
```
